### list vs array
 NumPy is:

Faster  
Uses less memory   
Supports vector operations  

```python
# Python list
a = [1, 2, 3]

# NumPy array
b = np.array([1, 2, 3])
```

### Shape & Size

```python
arr = np.array([[1,2,3],[4,5,6]])

print(arr.shape)  
print(arr.size)   
```

### Indexing & Slicing

```python
arr = np.array([10, 20, 30, 40])

print(arr[0])     
print(arr[1:3]) 
```

```python
arr = np.array([[1,2,3],[4,5,6]])

print(arr[1,0])  
```

### Array Operations

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)   
print(a * b)   
```

no loop needed here while in list doing operation needs loop but in array we don't need loops for this purpose, this is called **vectorization**

### Built-in Functions

```python
arr = np.array([1,2,3,4])

print(np.mean(arr)) 
print(np.sum(arr))   
print(np.max(arr))   
```
