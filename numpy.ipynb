### list vs array

Uses less memory

### Shape & Size

```python
arr = np.array([[1,2,3],[4,5,6]])

print(arr.shape)  
print(arr.size)   
```

### Indexing & Slicing

```python
arr = np.array([10, 20, 30, 40])

print(arr[0])     
print(arr[1:3]) 
```

```python
arr = np.array([[1,2,3],[4,5,6]])

print(arr[1,0])  
```

### Array Operations

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)   
print(a * b)   
```

no loop needed here while in list doing operation needs loop but in array we don't need loops for this purpose, this is called **vectorization**

### Built-in Functions

```python
arr = np.array([1,2,3,4])

print(np.mean(arr)) 
print(np.sum(arr))   
print(np.max(arr)) 
print(np.var(arr))  
print(np.std(arr))
```

### Special Arrays

```python
import numpy as np
a =np.zeros((2,2))
b =np.ones((3,3))
c =np.arange(0,10,2)
d =np.linspace(0,1,5)

print(a)
print(b)
print(c)
print(d)
```

### Reshaping Arrays

```python
arr = np.array([1,2,3,4,5,6])
print(arr)
new_arr = arr.reshape(2,3)
print(new_arr)
```

### Flattening

```python
arr = np.array([1,2,3,4,5,6])
arr.flatten()
```

### Random Numbers

```python
print(np.random.rand(3))
print(np.random.randint(1, 10, 5))
```

### Boolean Indexing

```python
arr = np.array([10, 20, 30, 40])
print(arr[arr > 20])  
```

### Broadcasting

```python
a = np.array([1,2,3])
b = 3
print(a * b)  
```

### Matrix Operations

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])

print(np.dot(a, b))
```

### Axis Concept

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr)
print(np.sum(arr, axis=0))  
print(np.sum(arr, axis=1))  
```

```python
cof = np.array([[1,1],[5,1]])
cons = np.array([3,1])
```
